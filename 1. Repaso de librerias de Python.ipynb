### **Carga y Creación**

#### Lista

- Es una colección ordenada de elementos.
- Puede contener distintos tipos de datos (números, strings, listas, etc.).
- Se puede modificar (mutable).

```python
# Crear una lista en Python
lista = [10, 20, 30, 40, 50]

# Acceso a elementos
print("Primer elemento:", lista[0])  # Acceder al primer elemento

# Modificar un elemento
lista[1] = 25
print("Lista modificada:", lista)

# Agregar un elemento al final
lista.append(60)

# Insertar en una posición específica
lista.insert(2, 100)

# Eliminar un elemento
lista.remove(40)

# Último valor eliminado
ultimo_valor = lista.pop()

print("Lista final:", lista)
print("Último valor eliminado:", ultimo_valor)
```

#### Arrays de NumPy

- Similar a una lista, pero más eficiente para operaciones matemáticas.
- Soporta operaciones vectorizadas sin necesidad de loops.
- Ideal para cálculos numéricos en ML/IA.

```python
# Crear un array de NumPy
array = np.array([1, 2, 3, 4, 5])

# Operaciones vectorizadas
print("Suma de todos los elementos:", array.sum())
print("Promedio:", array.mean())
print("Máximo:", array.max())
print("Mínimo:", array.min())

# Transformación del array
array_doble = array * 2  # Multiplica cada elemento por 2
print("Array Doble:", array_doble)

#array al cuadrado
array_cuadrado = array ** 2  # Multiplica cada elemento por 2
print("Array Cuadrado:", array_cuadrado)
```

¿Por qué usar NumPy en vez de las listas de Python?

```python
### Uso de Memoria
import sys
numerosLista = range(1000)
print('Uso de memoria utilizado una lista:', sys.getsizeof(numerosLista[1]) * len (numerosLista))
numerosNumPy = np.arange(1000)
print('Uso de memoria utilizado numPy:', numerosNumPy.itemsize * numerosNumPy.size)

### Velocidad
import time
tamanioLista = 1000000
lista1 = range(tamanioLista)
lista2 = range(tamanioLista)
numpy1 = np.arange(tamanioLista)
numpy2 = np.arange(tamanioLista)
inicio1 = time.time()
suma1 = [x + y for x, y in zip(lista1, lista2)]
final1 = time.time()
inicio2 = time.time()
suma2 = numpy1 + numpy2
final2 = time.time()
print('Tiempo utilizando listas', (final1 - inicio1) * 1000, 'seg.')
print('Tiempo utilizando numPy', (final2 - inicio2) * 1000, 'seg.')
```

#### Diccionario

- Almacena datos en formato clave-valor.
- Permite un acceso rápido a los valores a través de claves.

```python
# Crear un diccionario
diccionario = {"Nombre": ["Ana", "Juan", "Pedro"], "Edad": [25, 30, 35]}

# Acceso a valores
print("Edades:", diccionario["Edad"])
print(type(diccionario["Edad"]))

# Agregar un nuevo par clave-valor
diccionario["Ciudad"] = ["Santiago", "Buenos Aires", "Madrid"]

# Modificar un valor
diccionario["Edad"][1] = 31  # Cambia la edad de Juan

# Obtener solo las claves o los valores
print("Claves:", diccionario.keys())
print("Valores:", diccionario.values())

print("Diccionario final:\n", diccionario)
```

##### Detección de Outliers

Material Complementario: https://www.analyticsvidhya.com/blog/2021/05/feature-engineering-how-to-detect-and-remove-outliers-with-python-code

```python
# Cargar CSV con headers distintos
df = pd.read_csv(r"C:\Users\Tomás Rubinstein\Desktop\clases_ucema_2026\Clase 1 - Repaso Python y Librerias\adult.data", header=None,
                              names=["age", "workclass", "fnlwgt", "education",
                                     "education-num", "marital-status",
                                     "occupation", "relationship", "race", "sex",
                                     "capital-gain", "capital-loss",
                                     "hours-per-week","native-country", "income"]
                              )
df
```

```python
df.shape
```

```python
#Analizamos si la columna hours-per-week presenta outliers
plt.figure(figsize=(15,5))
#Primero vemos su distribución
sns.distplot(df['hours-per-week'])
```

```python
#Después utilizamos un gráfico de violín para chequear presencia de outliers
sns.boxplot(data=df,x=df['hours-per-week'])
```

```python
# Método IQR para detección de Outliers

# Calculamos el primer y tercer cuartil de la data de hours-per-week
Q1 = df['hours-per-week'].quantile(0.25)
Q3 = df['hours-per-week'].quantile(0.75)
print("El valor del primer cuartil es:", Q1, " y el del tercer cuartil es:", Q3)

#Calculamos los límites hacia abajo y arriba
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
print (lower)
print (upper)
```

```python
#Buscamos identificar aquellos valores que son upper outliers
upper_outliers = df[df['hours-per-week'] > upper].sort_values("hours-per-week", ascending = False)
upper_outliers
```

```python
#Buscamos identificar aquellos valores que son lower outliers
lower_outliers = df[df['hours-per-week'] < lower].sort_values("hours-per-week")
lower_outliers
```

```python
# Creamos un array con las filas que presentan outliers
upper_array = np.where(df['hours-per-week']>=upper)[0]
lower_array = np.where(df['hours-per-week']<=lower)[0]
```

```python
# Removemos los outliers
df.drop(index=upper_array, inplace=True)
df.drop(index=lower_array, inplace=True)
```

##### Rename(): Renombrar columnas o índices. [Documentación](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html)

```python
df = df.rename(columns = {0: "age"})
df.rename(columns={1: "workclass", 2: "fnlwgt", 3: "education", 4: "education-num",
    5: "marital-status", 6: "occupation", 7: "relationship", 8: "race", 9: "sex",
    10: "capital-gain", 11: "capital-loss", 12: "hours-per-week", 13: "native-country",
    14: "income"
}, inplace = True)
df.head(1)
```

```python
# Creamos la columna de minutos por semana
df["minutes-per-week"] = df["hours-per-week"].apply(lambda x: x*60)
df.head(3)
```

##### Replace(): Reemplazar un valor. [Documentación](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html)


```python
df["native-country"].unique()
```

```python
df["native-country"] = df["native-country"].replace(' ?', np.nan)
df["native-country"].unique()
```

#### Cambios de Tipo de Dato

```python
df.info()
```

```python
print(df["native-country"].unique())

#Elimino los espacios
df["native-country"] = df["native-country"].str.strip()

#Paso todo a minúscula
df["native-country"] = df["native-country"].str.lower()
print(df["native-country"].unique())
```

```python
#Lo hago para todas las columnas
for column in df.columns:
  if df[column].dtype == object:
    df[column] = df[column].str.strip()
    df[column] = df[column].str.lower()
df.head(3)
```

```python
#Paso de entero a float
df["capital-gain"] = df["capital-gain"].astype(float)
df.head(3)
```

```python
#Paso de Objeto a Booleano cambiando el nombre de la columna income
df["income"] = df["income"].map({"<=50k": 0, ">50k": 1})
df.rename(columns={"income": "more-50k"}, inplace = True)
df.head(3)
```

```python
#Agrego una columna de fecha con formato string y la paso a datetime
df["interview-date"] = "2022-01-01"
df["interview-date"] = pd.to_datetime(df["interview-date"])
df.head(3)
```

```python
#Agrego una columna de interview-year y otra de interview-month
df["interview-year"] = df["interview-date"].dt.year
df["interview-month"] = df["interview-date"].dt.month
df.head(3)
```

```python
print(df['workclass'].unique())
# Como en este caso considero a state-gov, federal-gov y local-gov como lo mismo,
# cambio todos los federal-gov y local-gov por state-gov
df['workclass'] = df['workclass'].replace(['federal-gov', 'local-gov'], 'state-gov')
df['workclass'].unique()
```

```python
print(df['race'].unique())
# Para utilizar modelos de ML con variables cualitativas, es necesario
# pasarlas a algo cuantitativo o dummie. Eso es lo que hacemos:
df = pd.get_dummies(df, columns=["race"], prefix="race")
df.head(1)
```

En este caso como sólo el 1.8% de nuestros datos son nulos, podríamos elegir eliminar estos valores (eliminando su fila).

```python
#Otros métodos de dropna

# Aplicamos la condición de que todos los valores del registro sean NaNs para borrar el registro.
df.dropna(how='all', inplace = True)

# O podemos aplicar un límite con un número mínimo de valores no NaNs por registro.
df = df.dropna(thresh=3)
```

2- Asignar un nuevo valor a los nulos

```python
df_2 = pd.DataFrame({'Nombre': ['Juan', np.nan, 'Susana',np.nan, 'Juan', 'Olga'],
                  'Edad': [20, 10, np.nan,np.nan, 20, 10],
                  'Peso': [74, 42, 85,60, 74, np.nan],
                  'Altura': [1.84, 1.54, np.nan,np.nan, 1.84, 1.54]
                 },
                 index=['Individuo-1', 'Individuo-2', 'Individuo-3', 'Individuo-4', 'Individuo-5', 'Individuo-6'])
```

```python
print(df_2.isnull().sum()/len(df_2))
```

```python
# Rellenamos los valores Nan de Edad con la media de edad
print(df_2['Edad'])
df_2['Edad'] = df_2['Edad'].fillna(np.mean(df_2['Edad']))
df_2['Edad']
```

```python
#Rellenamos los valores Nan de Altura usando el método bfill
print(df_2['Altura'])
df_2['Altura'] = df_2['Altura'].fillna(method='bfill')
df_2['Altura']
```

```python
#Rellenamos los valores Nan de Peso usando el valor de Peso del individuo más cercano a la Altura
print(df_2['Peso'])
df_2 = df_2.sort_values("Altura")
print(df_2)  # Ordenamos por altura
df_2["Peso"] = df_2["Peso"].fillna(method="ffill")
df_2['Peso']
```
