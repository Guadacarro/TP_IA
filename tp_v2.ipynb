Nos quedamos con el 25% de registros de cada mes de manera aleatoria. Agregamos un random_state para que nos de siempre la misma muestra de datos. El dataset es muy grande entonces necesitamos reducirlo para que no lleve tanto tiempo correr las APIs y luego quizás el modelo.

```python
df_2022 = df_2022.groupby('mes', group_keys=False).apply(
    lambda x: x.sample(frac=0.25, random_state=42)
)

print(f"Filas después de la muestra: {len(df_2022)}")
```

Filtramos por aeropuertos más importantes.

```python
top_aeropuertos = df_2022['origen'].value_counts()
cobertura = top_aeropuertos.cumsum() / top_aeropuertos.sum()
aeropuertos_relevantes = cobertura[cobertura <= 0.90].index.tolist()

df_2022 = df_2022[df_2022['origen'].isin(aeropuertos_relevantes)]

print(f"Filas después de filtrar aeropuertos: {len(df_2022)}")
print(f"Cantidad de aeropuertos relevantes: {len(aeropuertos_relevantes)}")
```

Hacemos un llamado a una API para obtener la longitud y latitud de cada aeropuerto.

```python
import requests

airports_db = pd.read_csv("https://ourairports.com/data/airports.csv")
airports_db = airports_db[airports_db['iata_code'].isin(aeropuertos_relevantes)]
airports_db = airports_db[['iata_code', 'latitude_deg', 'longitude_deg']].dropna()
airports_db = airports_db.rename(columns={'iata_code': 'origen', 'latitude_deg': 'lat', 'longitude_deg': 'lon'})

print(f"Aeropuertos con coordenadas encontradas: {len(airports_db)}")
airports_db.head()
```

Con esa longitud y latitud que conseguimos, podemos obtener el clima de los registros que pertencen a nuestra dataset, en determinado momento en el aeropuerto de origen.

```python
import time

def clima_por_aeropuerto():
    clima_data = []

    for i, row in airports_db.iterrows():
        try:
            response = requests.get(
                "https://archive-api.open-meteo.com/v1/archive",
                params={
                    "latitude": row['lat'],
                    "longitude": row['lon'],
                    "start_date": "2022-01-01",
                    "end_date": "2022-12-31",
                    "daily": "temperature_2m_mean,precipitation_sum,windspeed_10m_max",
                    "timezone": "America/New_York"
                },
                timeout=30
            )
            data = response.json()
            temp = pd.DataFrame({
                'fecha_vuelo': data['daily']['time'],
                'temperatura_media': data['daily']['temperature_2m_mean'],
                'precipitacion': data['daily']['precipitation_sum'],
                'viento_max': data['daily']['windspeed_10m_max']
            })
            temp['origen'] = row['origen']
            clima_data.append(temp)
            print(f"OK: {row['origen']} ({i+1}/{len(airports_db)})")
        except Exception as e:
            print(f"Error con {row['origen']}: {e}")
        time.sleep(0.2)

    clima_data = pd.concat(clima_data, ignore_index=True)
    clima_data['fecha_vuelo'] = pd.to_datetime(clima_data['fecha_vuelo'])

    print('Guardamos el parquet de climas del año de cada aeropuerto...')
    clima_data.to_parquet('clima_2022_por_aeropuerto.parquet', index=False)
    print('Archivo guardado: clima_2022_por_aeropuerto.parquet')

    return clima_data
```

```python
if not os.path.exists('clima_2022_por_aeropuerto.parquet'):
   print('El archivo clima_2022_por_aeropuerto.parquet no existe. Comienza el llamado a la api para armar el parquet...')
   clima_data = clima_por_aeropuerto()
else:
   print('El archivo clima_2022_por_aeropuerto.parquet si existe.')
   clima_data = pd.read_parquet('clima_2022_por_aeropuerto.parquet')


print('clima_2022_por_aeropuerto.parquet cargado!')
print(f"\nFilas de clima obtenidas: {len(clima_data)}")
clima_data.head()
```

#### Análisis Outliers

```python
# Detección de outliers por el método IQR 
resultados = []
for col in numerical_columns:
    Q1 = df_2022[col].quantile(0.25)
    Q3 = df_2022[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_2022[(df_2022[col] < lower_bound) | (df_2022[col] > upper_bound)]
    if len(outliers) > 0:  # nos quedamos solo con las que tienen outliers
        resultados.append({
            'variable': col,
            'cant_outliers': len(outliers),
            'porcentaje': round(len(outliers) / len(df_2022) * 100, 2),
        })

outliers_resumen = (pd.DataFrame(resultados)
                      .sort_values('cant_outliers', ascending=False)
                      .reset_index(drop=True))

print(f"{len(outliers_resumen)} de {len(numerical_columns)} variables numéricas tienen outliers:\n")
print(outliers_resumen.to_string(index=False))
```

```python
def plot_categoricas(columnas, titulo, df, top_n=10):
    fig, axes = plt.subplots(1, len(columnas), figsize=(9 * len(columnas), 5))
    if len(columnas) == 1:
        axes = [axes]
    for i, col in enumerate(columnas):
        top = df[col].value_counts().nlargest(top_n)
        total = top.sum()
        axes[i].barh(top.index[::-1], top.values[::-1], color=sns.color_palette('viridis_r', len(top)))
        axes[i].set_title(f'Top {top_n} - {col}', fontsize=11)
        for bar in axes[i].patches:
            percentage = '{:.1f}%'.format(100 * bar.get_width() / total)
            axes[i].annotate(percentage, (bar.get_width(), bar.get_y() + bar.get_height() / 2),
                             ha='left', va='center', fontsize=8, xytext=(5, 0), textcoords='offset points')
    plt.suptitle(titulo, fontsize=13)
    plt.tight_layout()
    plt.show()

plot_categoricas(['aerolinea', 'codigo_cancelacion'], 'Distribuciones categóricas - Aerolínea y Cancelación', df_2022)
plot_categoricas(['origen', 'destino'], 'Distribuciones categóricas - Aeropuertos origen y destino', df_2022)
plot_categoricas(['ciudad_origen', 'ciudad_destino'], 'Distribuciones categóricas - Ciudades origen y destino', df_2022)
```

```python
def resumen_por_variable(df, var, orden=None):
    resumen = df.groupby(var, observed=True).agg(
        cant_vuelos=('retrasado', 'size'),
        tasa_retraso=('retrasado', 'mean')
    ).reset_index()
    if orden is not None:
        resumen[var] = pd.Categorical(resumen[var], categories=orden, ordered=True)
        resumen = resumen.sort_values(var)
    else:
        resumen = resumen.sort_values(var)
    return resumen

orden_dias = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

resumen_hora = resumen_por_variable(df_2022, 'hora_prog')
resumen_dia = resumen_por_variable(df_2022, 'nombre_dia', orden=orden_dias)
resumen_mes = resumen_por_variable(df_2022, 'mes')

print(resumen_hora)
print(resumen_dia)
print(resumen_mes)
```

```python
def plot_volumen_vs_retraso(resumen, var, titulo):
    fig, ax1 = plt.subplots(figsize=(10,6))

    ax1.bar(resumen[var].astype(str), resumen['cant_vuelos'], color='steelblue', alpha=0.6)
    ax1.set_ylabel('Cantidad de vuelos', color='steelblue')
    ax1.tick_params(axis='x', rotation=45)

    ax2 = ax1.twinx()
    ax2.plot(resumen[var].astype(str), resumen['tasa_retraso'], color='red', marker='o')
    ax2.set_ylabel('Tasa de retraso', color='red')

    plt.title(titulo)
    fig.tight_layout()
    plt.show()

plot_volumen_vs_retraso(resumen_hora, 'hora_prog', 'Volumen de vuelos vs tasa de retraso por hora')
plot_volumen_vs_retraso(resumen_dia, 'nombre_dia', 'Volumen de vuelos vs tasa de retraso por día')
plot_volumen_vs_retraso(resumen_mes, 'mes', 'Volumen de vuelos vs tasa de retraso por mes')
```

Features históricas (lag features): calculamos métricas basadas en vuelos anteriores. No es leakage porque solo usamos datos del pasado.

```python
abt = abt.sort_values('fecha_vuelo').reset_index(drop=True)

# 1. Tasa de retraso por aerolínea en los últimos 7 días
diario_aero = (abt.groupby(['codigo_aerolinea', 'fecha_vuelo'])['retrasado']
               .mean().reset_index(name='_tasa_dia_aero'))
diario_aero['tasa_retraso_aerolinea_7d'] = (
    diario_aero.groupby('codigo_aerolinea')['_tasa_dia_aero']
    .transform(lambda x: x.shift(1).rolling(7, min_periods=1).mean())
)
abt = abt.merge(diario_aero[['codigo_aerolinea', 'fecha_vuelo', 'tasa_retraso_aerolinea_7d']],
                on=['codigo_aerolinea', 'fecha_vuelo'], how='left')

# 2. Tasa de retraso por aeropuerto origen en los últimos 7 días
diario_orig = (abt.groupby(['origen', 'fecha_vuelo'])['retrasado']
               .mean().reset_index(name='_tasa_dia_orig'))
diario_orig['tasa_retraso_origen_7d'] = (
    diario_orig.groupby('origen')['_tasa_dia_orig']
    .transform(lambda x: x.shift(1).rolling(7, min_periods=1).mean())
)
abt = abt.merge(diario_orig[['origen', 'fecha_vuelo', 'tasa_retraso_origen_7d']],
                on=['origen', 'fecha_vuelo'], how='left')

# 3. Cantidad de vuelos ese día en el aeropuerto origen (proxy de congestión)
vuelos_dia = abt.groupby(['origen', 'fecha_vuelo']).size().reset_index(name='vuelos_dia_origen')
abt = abt.merge(vuelos_dia, on=['origen', 'fecha_vuelo'], how='left')

# 4. Si el vuelo anterior en la misma ruta tuvo retraso (efecto cascada)
abt['retraso_previo_ruta'] = (
    abt.groupby(abt['origen'] + '-' + abt['destino'])['retrasado']
    .shift(1).fillna(0).astype(int)
)

# Rellenar NaN de los primeros días sin historial con media acumulada (expanding) para evitar leakage 
media_acumulada = abt['retrasado'].expanding().mean()
abt['tasa_retraso_aerolinea_7d'] = abt['tasa_retraso_aerolinea_7d'].fillna(media_acumulada)
abt['tasa_retraso_origen_7d'] = abt['tasa_retraso_origen_7d'].fillna(media_acumulada)

print("Features históricas creadas:")
abt[['tasa_retraso_aerolinea_7d', 'tasa_retraso_origen_7d', 'vuelos_dia_origen', 'retraso_previo_ruta']].describe().round(3)
```

## Modelo

### Definición del problema y métrica principal

El objetivo de este proyecto es construir un **clasificador binario** que prediga si un vuelo se va a retrasar más de 15 minutos (`retrasado = 1`) o no (`retrasado = 0`), utilizando únicamente información disponible **antes** de que el vuelo despegue.

**¿Por qué priorizamos Recall?**

El escenario de negocio es una aerolínea que quiere **alertar a sus pasajeros** ante posibles retrasos. Los costos de cada tipo de error son asimétricos:

| Error | Significado | Costo |
|-------|-------------|-------|
| **Falso Negativo** (no alertar un retraso real) | El pasajero no fue avisado, pierde conexiones, se frustra | **Alto** — pérdida de confianza y retención |
| **Falso Positivo** (alertar un retraso que no ocurre) | El pasajero recibe una alerta innecesaria | **Bajo** — molestia menor |

Cuando el costo del falso negativo supera al del falso positivo, la métrica correcta es **Recall**: ¿de todos los vuelos que realmente se retrasan, cuántos logra detectar el modelo?

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, classification_report,
    confusion_matrix, roc_curve, auc
)

import warnings
warnings.filterwarnings('ignore')
```

```python
print("Distribución de clases:")
print(abt_final['retrasado'].value_counts())
print()
print("Distribución en porcentaje:")
print(abt_final['retrasado'].value_counts(normalize=True).mul(100).round(2))
```

Existe un **desbalanceo de clases** moderado:

- Clase 0 (vuelo no retrasado): ~79%
- Clase 1 (vuelo retrasado): ~21%

Un modelo que prediga siempre "no retrasado" tendría ~79% de accuracy sin haber aprendido nada útil. Por eso el accuracy solo no es una métrica confiable en este escenario.

### Separación train / test

```python
X = abt_final.drop(columns=['retrasado'])
y = abt_final['retrasado']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42,
    stratify=y
)

print(f"Features: {X.shape[1]} columnas")
print(f"Train: {X_train.shape[0]} filas ({y_train.mean():.1%} retrasados)")
print(f"Test:  {X_test.shape[0]} filas ({y_test.mean():.1%} retrasados)")
```

- **70% train / 30% test**: proporción estándar que deja suficientes datos para entrenar y evaluar.
- **`stratify=y`**: garantiza que ambos conjuntos mantengan la misma proporción de clases (~79/21) que el dataset original.
- **`random_state=42`**: hace que la división sea reproducible.

```python
def reportar_metricas(nombre, y_true, y_pred):
    return {
        'Modelo': nombre,
        'Accuracy': round(accuracy_score(y_true, y_pred), 3),
        'Precision': round(precision_score(y_true, y_pred, zero_division=0), 3),
        'Recall': round(recall_score(y_true, y_pred, zero_division=0), 3),
        'F1': round(f1_score(y_true, y_pred, zero_division=0), 3),
        'ROC-AUC': round(roc_auc_score(y_true, y_pred), 3),
    }
```

### Modelo 1: Regresión Logística (baseline)

La Regresión Logística es un modelo lineal de clasificación. Es un buen punto de partida porque:
- Es simple e interpretable.
- Permite visualizar conceptos como **regularización** y **sesgo/varianza**.

Usamos un **Pipeline** que encadena `StandardScaler` (escalado) + `LogisticRegression` en un solo objeto. Esto garantiza que el escalado se ajuste solo con datos de train y se aplique consistentemente al test, evitando data leakage.

```python
pipe_base = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(random_state=42, max_iter=1000))
])

pipe_base.fit(X_train, y_train)
y_pred_train_base = pipe_base.predict(X_train)
y_pred_test_base = pipe_base.predict(X_test)

print("=== Regresión Logística SIN balanceo ===")
print(classification_report(y_test, y_pred_test_base, target_names=['No retrasado', 'Retrasado']))
```

**Resultado sin balanceo:** el accuracy es alto (~79%), pero el recall de la clase "Retrasado" es extremadamente bajo (~5%). El modelo aprendió la estrategia "segura" de predecir siempre "no retrasado" porque estadísticamente acierta la mayoría de las veces, sin haber aprendido qué causa los retrasos.

Esto demuestra por qué **el accuracy es engañoso con clases desbalanceadas**.

```python
pipe_balanced = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(
        class_weight='balanced',
        random_state=42,
        max_iter=1000
    ))
])

pipe_balanced.fit(X_train, y_train)
y_pred_train_bal = pipe_balanced.predict(X_train)
y_pred_test_bal = pipe_balanced.predict(X_test)

print("=== Regresión Logística CON balanceo ===")
print(classification_report(y_test, y_pred_test_bal, target_names=['No retrasado', 'Retrasado']))
```

`class_weight='balanced'` ajusta los pesos de la función de pérdida para que los errores sobre la clase minoritaria (retrasados) se penalicen proporcionalmente más. Esto fuerza al modelo a prestarle atención a esa clase en lugar de ignorarla.

**Resultado con balanceo:** el accuracy baja, pero el recall se dispara significativamente. El modelo ahora sí detecta vuelos retrasados.

```python
resultados_lr = pd.DataFrame([
    reportar_metricas('RL sin balanceo', y_test, y_pred_test_base),
    reportar_metricas('RL balanceada', y_test, y_pred_test_bal),
])

metricas_cols = ['Accuracy', 'Precision', 'Recall', 'F1', 'ROC-AUC']
x = np.arange(len(metricas_cols))
ancho = 0.35

fig, ax = plt.subplots(figsize=(10, 5))
for i, (_, row) in enumerate(resultados_lr.iterrows()):
    color = 'gray' if i == 0 else 'steelblue'
    valores = [row[m] for m in metricas_cols]
    bars = ax.bar(x + i * ancho - ancho/2, valores, ancho, label=row['Modelo'], color=color)
    for bar in bars:
        h = bar.get_height()
        ax.annotate(f'{h:.3f}', xy=(bar.get_x() + bar.get_width()/2, h),
                    xytext=(0, 3), textcoords='offset points', ha='center', fontsize=9)

ax.set_xticks(x)
ax.set_xticklabels(metricas_cols)
ax.set_ylim(0, 1)
ax.set_ylabel('Score')
ax.set_title('Regresión Logística: sin balanceo vs balanceada')
ax.legend()
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Análisis sesgo / varianza — Regresión Logística

Comparar las métricas en **train vs test** nos permite diagnosticar problemas de sesgo y varianza:

| Situación | Train | Test | Diagnóstico |
|-----------|-------|------|-------------|
| Train alto, test bajo | ✅ | ❌ | **Overfitting** (alta varianza) — el modelo memoriza el train pero no generaliza |
| Ambos bajos | ❌ | ❌ | **Underfitting** (alto sesgo) — el modelo es demasiado simple |
| Ambos similares y razonables | ✅ | ✅ | Buen balance |

```python
print("=== Sesgo/Varianza — Regresión Logística ===")
print()
print("SIN balanceo:")
print(f"  Recall Train: {recall_score(y_train, y_pred_train_base):.3f}")
print(f"  Recall Test:  {recall_score(y_test, y_pred_test_base):.3f}")
print()
print("CON balanceo:")
print(f"  Recall Train: {recall_score(y_train, y_pred_train_bal):.3f}")
print(f"  Recall Test:  {recall_score(y_test, y_pred_test_bal):.3f}")
print()
print("Accuracy Train (bal):", round(accuracy_score(y_train, y_pred_train_bal), 3))
print("Accuracy Test  (bal):", round(accuracy_score(y_test, y_pred_test_bal), 3))
```

La Regresión Logística con balanceo muestra métricas similares entre train y test, lo que indica que **no hay overfitting significativo**. Esto es esperable: la Regresión Logística es un modelo lineal con **regularización L2 por defecto** (parámetro `C=1.0`), lo que limita la complejidad del modelo y previene el sobreajuste.

> **Nota sobre regularización:** en sklearn, el parámetro `C` es el inverso de la fuerza de regularización. Un `C` más bajo → más regularización → modelo más simple. El valor por defecto `C=1.0` ofrece un balance razonable.

### Modelo 2: XGBoost (boosting)

XGBoost es un algoritmo de ensamble basado en **boosting**: construye árboles de decisión de forma **secuencial**, donde cada nuevo árbol se enfoca en corregir los errores que cometieron los anteriores.

A diferencia del **bagging** (como Random Forest, que entrena árboles independientes en paralelo y promedia sus predicciones), el boosting **aprende de manera incremental**, lo que suele lograr mejor precisión a costa de mayor riesgo de overfitting si no se regulariza adecuadamente.

```python
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()
print(f"scale_pos_weight: {scale_pos_weight:.2f}")
print(f"  → Clase 0: {(y_train == 0).sum()} | Clase 1: {(y_train == 1).sum()}")
print(f"  → Ratio: por cada vuelo retrasado hay {scale_pos_weight:.1f} vuelos a tiempo")
```

XGBoost no tiene el parámetro `class_weight`. En su lugar usa `scale_pos_weight`, que cumple la misma función: le indica al modelo cuántas veces más debe pesar un error en la clase positiva (retrasado) respecto a la negativa.

```python
param_dist = {
    'n_estimators': [100, 150, 200, 250],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'scale_pos_weight': [1, 2, scale_pos_weight],
    'gamma': [0, 1, 2, 3],
    'subsample': [0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.7, 0.8, 0.9, 1.0],
    'reg_alpha': [0, 0.1, 1],
    'reg_lambda': [1, 2, 5],
}

total_combinaciones = 1
for v in param_dist.values():
    total_combinaciones *= len(v)
print(f"Espacio total de búsqueda: {total_combinaciones:,} combinaciones")
print(f"Se probarán 50 combinaciones al azar (RandomizedSearchCV)")
```

**Hiperparámetros y regularización:**

| Parámetro | Qué controla | Efecto de regularización |
|-----------|-------------|------------------------|
| `max_depth` | Profundidad máxima de cada árbol | Valores bajos → árboles simples → menos overfitting |
| `gamma` | Reducción mínima de pérdida para hacer un split | Valores altos → solo splits que aporten mucho |
| `reg_alpha` | Regularización L1 (penaliza cantidad de features) | Fuerza sparsity, selecciona features |
| `reg_lambda` | Regularización L2 (penaliza magnitud de pesos) | Suaviza las predicciones |
| `subsample` | Fracción de datos usados por árbol | < 1.0 introduce aleatoriedad → reduce varianza |
| `colsample_bytree` | Fracción de features usadas por árbol | Similar a subsample pero sobre columnas |

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

xgb_model = XGBClassifier(
    random_state=42,
    eval_metric='logloss',
    n_jobs=-1
)

random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=50,
    scoring='recall',
    cv=cv,
    random_state=42,
    n_jobs=-1,
    verbose=1
)

random_search.fit(X_train, y_train)

print(f"Mejor Recall (CV): {random_search.best_score_:.3f}")
print()
print("Mejores hiperparámetros:")
for param, valor in random_search.best_params_.items():
    print(f"  {param}: {valor}")

mejor_xgb = random_search.best_estimator_
```

**¿Qué es la validación cruzada (cross-validation)?**

Usamos `StratifiedKFold` con 5 folds: el set de train se divide en 5 partes iguales, manteniendo la proporción de clases en cada una. El modelo se entrena en 4 partes y se evalúa en la 5ta, rotando 5 veces. Esto da una estimación más robusta del rendimiento porque **cada dato es usado exactamente una vez para evaluación**.

`RandomizedSearchCV` combina esto con una búsqueda aleatoria de hiperparámetros: prueba 50 combinaciones al azar del espacio de búsqueda y se queda con la que logra el mejor recall promedio en los 5 folds.

```python
y_pred_train_xgb = mejor_xgb.predict(X_train)
y_pred_test_xgb = mejor_xgb.predict(X_test)

print("=== XGBoost optimizado ===")
print(classification_report(y_test, y_pred_test_xgb, target_names=['No retrasado', 'Retrasado']))

print("--- Sesgo/Varianza ---")
print(f"Recall  Train: {recall_score(y_train, y_pred_train_xgb):.3f}")
print(f"Recall  Test:  {recall_score(y_test, y_pred_test_xgb):.3f}")
print(f"Accuracy Train: {accuracy_score(y_train, y_pred_train_xgb):.3f}")
print(f"Accuracy Test:  {accuracy_score(y_test, y_pred_test_xgb):.3f}")
```

Si el recall en train es significativamente mayor que en test, hay indicios de **overfitting**. Si ambos son similares, el modelo generaliza bien. La regularización de XGBoost (`gamma`, `max_depth`, `reg_alpha`, `reg_lambda`) ayuda a controlar este balance.

### Ajuste del umbral de decisión

```python
y_proba_xgb = mejor_xgb.predict_proba(X_test)[:, 1]

umbrales = [0.30, 0.35, 0.40, 0.45, 0.50, 0.55]
resultados_umbral = []
for thr in umbrales:
    y_pred_thr = (y_proba_xgb >= thr).astype(int)
    resultados_umbral.append({
        'Umbral': thr,
        'Accuracy': round(accuracy_score(y_test, y_pred_thr), 3),
        'Precision': round(precision_score(y_test, y_pred_thr, zero_division=0), 3),
        'Recall': round(recall_score(y_test, y_pred_thr, zero_division=0), 3),
        'F1': round(f1_score(y_test, y_pred_thr, zero_division=0), 3),
    })

df_umbrales = pd.DataFrame(resultados_umbral)
df_umbrales.style.highlight_max(subset=['Recall'], color='lightgreen').highlight_max(subset=['F1'], color='lightyellow')
```

Por defecto, un modelo clasifica como positivo si la probabilidad predicha es ≥ 0.50. Pero este umbral no siempre es óptimo.

Al **bajar el umbral a 0.40**, el modelo se vuelve más "sensible": marca más vuelos como retrasados, lo que aumenta el recall (detecta más retrasos reales) a costa de bajar la precision (más falsas alarmas).

Elegimos el **umbral 0.40** porque:
- De cada 10 vuelos que el modelo marca como retrasados, ~3 efectivamente se retrasan (precision ~29%).
- De cada 10 vuelos que realmente se retrasan, el modelo detecta ~8 (recall ~79%).
- Este trade-off es aceptable dado el objetivo de negocio: **el costo de no alertar un retraso real es mayor que el de emitir una falsa alarma**.

### Evaluación final

A lo largo del trabajo se evaluaron dos modelos de clasificación para predecir si un vuelo se retrasaría más de 15 minutos:

1. **Regresión Logística** con `class_weight='balanced'` (baseline)
2. **XGBoost** optimizado mediante `RandomizedSearchCV` con cross-validation

**Modelo seleccionado: XGBoost con umbral de decisión ajustado a 0.40.**

| Aspecto | Detalle |
|---------|---------|
| **Métrica priorizada** | Recall, por el objetivo de negocio (detectar retrasos para alertar al pasajero) |
| **Recall final** | ~79% — detecta correctamente ~8 de cada 10 vuelos que realmente se retrasan |
| **Precision** | ~29% — de las alertas emitidas, ~3 de cada 10 corresponden a retrasos reales |
| **Trade-off aceptado** | Mayor tasa de falsas alarmas a cambio de minimizar retrasos no detectados |

**Conceptos aplicados:**
- **Sesgo/varianza:** se evaluó comparando métricas en train vs test para detectar overfitting
- **Regularización:** tanto en la Regresión Logística (L2 vía parámetro `C`) como en XGBoost (`gamma`, `max_depth`, `reg_alpha`, `reg_lambda`)
- **Cross-validation:** `StratifiedKFold` con 5 folds para optimización robusta de hiperparámetros
- **Boosting:** XGBoost como algoritmo de ensamble secuencial

El modelo final logra un balance razonable entre detección de retrasos y credibilidad de las alertas, en línea con el criterio de negocio definido: **el costo de no anticipar un retraso real al cliente es mayor al de emitir una alerta que finalmente no se concreta**.
